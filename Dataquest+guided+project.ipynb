# Factorials and Combinations functions

our goal is to write code that can enable users to answer probability questions about playing the lottery. Throughout the project, we will need to calculate repeatedly probabilities and combinations. As a consequence, we will start by wrting two functions that we will use often:
- A function that calculates factorials 
- A function that calculates combinations.
To calculate factorials, this is the formular we learned we need to use:
 `n! = n * (n-1)*(n-2)*...2*1` 
In the 6/49 lottery, six numbers are drawwn from a set of 49 numbers that range from 1 to 49. The drawing is done without replacement, which means once a number is drawn, it is not put back in the set.

To find the number of combinations when we are sampling without replacement and taking only k objects from a group of n objects, we can use the formular:
`nCk = n!/k!(n-k)!

```python
def factorial(n):
    outcome = 1
    for i in range(1,n):
        outcome = outcome * (n-i)
    final_outcome = n * outcome
    return final_outcome
factorial(4)
```

```python
def combinations(n,k):
    if k <n and k!=0:
        n_factorial = factorial(n)
        denominator = factorial(n-k) * factorial(k)
        return n_factorial/denominator
    elif n ==k:
        return factorial(n)/factorial(k)
    elif k==0:
        return factorial(n)/factorial(n-k)
combinations(4,0)
```

In the 6/49 lottery, six numbers are drawn from a set 49 numbers that range from 1 to 49. A player wins the big prize if the six numbers on their tickets match all the six numbers drawn. If a player has a ticket with the numbers {13,22,24,27,42,44},he only wins the big prize if the numbers drawn are {13,22,24,27,42,44}. If only one number differs,he dose not win.

For the first version of the app, we want players to be able to calculate the probability of winning the big prize with various numbers they play on a single ticket (for rach ticket a player chooses six numbers out of 49). So we will start by building a function that calculates the probability of winning the big prize for any given ticket.

We discussed with the engineering team of the mediacl institue and they told us we need to be aware of the following details when we write the function:
- Inside the app, the user inputs six different numbers from 1 to 49
- Under the hood, the six numbers will come as a python list, which will serve as the single input to our function.
- The engineering team wants the function to print the probability value in a friendly way - in a way that people without any probability training are able to understand

```python
def one_ticket_probability(n):
    outcomes = combinations(49,len(n))
    successful_outcome = 1
    probability = successful_outcome/outcomes
    probability_percentage = probability *100
    return_value = "Your probability of winning with this number {} are {:.7}%. In other words you have a chance of 1 in {:,} chances to win".format(n,probability_percentage,combinations(49,len(n)))    
    return return_value 
```

```python
def multi_ticket_probability(n_tickets):
    total_outcomes = combinations(49,6)
    probability_percent = (n_tickets/total_outcomes)*100
    if n_tickets ==1:
        print('''Your chances to win the big prize with one ticket \n 
        is {:.6f}%. In other words, you have a 1 in  {:,} chances to win.'''.format(probability_percent,total_outcomes))
    else:
        combinations_simplified = round(n_tickets/total_outcomes)
        print('''Your chances to win the big prize with {} tickets \n
        is {:.6f}%. In other words, you have a 1 in \n
        {:,} chances to win'''.format(n_tickets,probability_percent,combinations_simplified))
```

```python
def probability_less_6(no_winning_number):
    """ no_winning_number must be lesser than 6 i.e between 2-5"""
    first_outcome = combinations(6,no_winning_number)
    second_outcome = combinations(43,6-no_winning_number)
    successful_outcome = first_outcome * second_outcome 
    total_possible_outcome = combinations(49,6)
    probability = successful_outcome/total_possible_outcome
    probability_percentage = probability * 100
    combination_simplified = round(total_possible_outcome/successful_outcome)
    print("""Your chance of winning with {} tickets is {:.6}%. \n
    That means you have 1 in {} """.format(no_winning_number,probability_percentage,combination_simplified))
    
```

```python
for i in [2,3,4,5]:
    probability_less_6(i)
```

```python
def probability_exactly_winning_no(winning_no):
    first_outcome = combinations(6,winning_no)
    second_outcome =  combinations(43,6-winning_no)
    successful_outcome = first_outcome*second_outcome
    possible_outcomes = combinations(49,6)
    probability = successful_outcome/possible_outcomes
    return probability
```

```python
probability_exactly_winning_no(2) *100
```

```python
def probability_less_than_winningno(winning_no):
    sum_probability = 0
    successful_outcome =0
    for i in range(winning_no,7):
        sum_probability += probability_exactly_winning_no(i) # starts from winning number and increases it
        successful_outcome +=(probability_exactly_winning_no(i)* combinations(49,6))
    combinations_simplified = round(combinations(49,6)/successful_outcome)
    print("""Your chances of winning at least {} winning numbers\n
    are {:.6}%. That means you have 1 in {}\n
    """.format(winning_no,sum_probability,combinations_simplified))   
```
